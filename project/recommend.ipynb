```python
import pandas as pd
import sqlite3
```

### user data preprocessing

```python
con = sqlite3.connect("db.sqlite3")
user = pd.read_sql("SELECT * FROM accounts_user", con, index_col=None)
```

```python
user = user.drop(['password', 'last_login', 'is_superuser', 'is_staff', 'username', 'first_name', 'last_name', 'email', 'nickname', 'date_joined'], axis=1)
user
```

### opensainvgs, opendeposits 더미 데이터 생성

```python
from random import randint, randrange

num = randint(1, 99)
print(num)
```

```python
user[user['id'] == num].income
balance = randrange(1000, int(user[user['id'] == num].income), 1000)
print(balance)
```

```python
pk = 1

new_list = []
while len(new_list) < 100:
    new_data = {"model": "accounts.opensaving", 'pk': pk}
    new_data["fields"] = {
        'user_id': randint(1, 99),
        'saving_id': randint(1, 63),
        'balance': randrange(1000, int(user[user['id'] == num].income), 1000)
    }
    new_list.append(new_data)
    pk += 1
```

```python
new_list
```

```python
import json

with open('opensavings.json', 'w', encoding='UTF-8') as m:
    json.dump(new_list, m, ensure_ascii=False, indent=4)
```
